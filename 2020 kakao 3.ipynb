```python
import copy

def rotate(key):
    new = []
    for x in zip(*key):
        new.append(list(reversed(x)))
    return new

def check(newlock,kn,ln):
    for i in range(ln):
        for j in range(ln):
            if newlock[kn-1+i][kn-1+j] != 1:
                return False
    return True

def keyin(newlock, key,r,c):
    ret = copy.deepcopy(newlock)    
    for i in range(len(key)):
        for j in range(len(key)):
            ret[r+i][c+j] += key[i][j]
    return ret    
```

```python
def solution(key, lock):
    kn = len(key)
    ln = len(lock)
    n = (kn-1)*2+len(lock)
    newlock = [[0]*n for _ in range(n)]
    for i in range(ln):
        for j in range(ln):
            newlock[kn-1+i][kn-1+j] = lock[i][j]
    
    for k in range(4):        
        for i in range(kn+ln-1):
            for j in range(kn+ln-1):
                tmp = keyin(newlock, key,i,j)
                print(k)
                for x in tmp:
                    print(x)
                if check(tmp, kn, ln):
                    return True
        
        key = rotate(key)
        
    
    return False
```
