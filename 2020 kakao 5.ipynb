```python
def isRight(r,c,typ,work):
    if typ == 0:
        if r==0 or work[r-1][c]==1 or work[r][c-1] == 2:
            return True
    elif typ == 1: #보
        if work[r-1][c] == 1:
            return True
        elif work[r-1][c+1] ==1:
            return True
        elif work[r][c-1]==2 and work[r][c+1]==2:
            return True
    return False

def check(r,c,typ,work,n):    
    work[r][c] = 0        
    for i in range(n+1):
        for j in range(n+1):
            if work[i][j]>0:
                if not isRight(i,j,work[i][j]-1, work):
                    work[r][c] = typ+1
                    return False
    return True


def solution(n, build_frame):
    work = [[0]*(n+1) for _ in range(n+1)]
    for c,r,typ,setup in build_frame:
        if setup == 1:
            if isRight(r,c,typ,work):
                work[r][c] = typ+1
        elif setup == 0:
            check(r,c,typ,work,n)       
          
    answer = []
    for i in range(n+1):
        for j in range(n+1):
            if work[i][j] >0:
                answer.append([j,i,work[i][j]-1])        
    answer.sort()
    return answer
```
