```python
import sys
from collections import deque
```

```python
n = int(sys.stdin.readline())
m = int(sys.stdin.readline())

dic = {i : set() for i in range(1,501)}
que = deque()
done = set()
for i in range(m):
    key, val = map(int, sys.stdin.readline().split())
    dic[key].add(val)
    dic[val].add(key)
que.append((1, 0))
done.add(1)
count = 0
while que:
    f, v = que.popleft()
    for p in dic[f]:
        if not p in done:
            que.append((p, v+1))
            done.add(p)
            if v+1 < 3:
                count +=1
print(count)
```
